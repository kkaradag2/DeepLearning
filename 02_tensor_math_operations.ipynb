```python
import torch
import numpy as np

shape=(2,3);
rand_tensor = torch.rand(shape);
ones_tensor = torch.ones(shape);
zeros_tensor = torch.zeros(shape);

print(rand_tensor);
print(ones_tensor);
print(zeros_tensor);
```

```python
tensor_a = torch.tensor([[2,4],
                        [5,8]]);


tensor_b = torch.tensor([[3,6],
                        [7,9]]);
# tensotleri birbirine eklemek için cat kullanılır
# dim = 0 satırdan ekler
tensor_c = torch.cat((tensor_a, tensor_b), dim=0);

# dim=1 sutundan ekle
tensor_d = torch.cat((tensor_a, tensor_b), dim=1);

print(tensor_c);

print(tensor_d);
```

```python
# birden fazla tensor de ekelenebilir
tensor_a = torch.rand(2,3);
tensor_b = torch.rand(2,3);
tensor_c = torch.rand(2,3);

# sutundan ekliyoruz, 4x3 matris oluyor
tensor_d= torch.cat((tensor_a, tensor_b, tensor_c), dim=0);

print(tensor_d);
```

# iki tensorun çarpımı

tensorleri birbirleri il eçarpman için torch.mul kullanılır

```python
# tek boyutlu
tensor_a = torch.tensor([1,2,3,4,5]);
tensor_b = torch.tensor([4,5,6,7,9]);

tensor_c = tensor_a.mul(tensor_b);  # axb

print(tensor_a);
print(tensor_b);
print(tensor_c);

```

```python
# 2x2 çarpımı

a= torch.tensor([[1,2],
                [3,4]]);

b= torch.tensor([[4,5],
                [6,7]]);

c= torch.mul(a,b);

print(a);
print(b);
print(c);
```

**Farklı Boyutlarda Tensör Çarpımı**

**torch.mul** fonksiyonu eleman-eleman çarpma yapar.
Bu nedenle yalnızca aynı boyutlu veya broadcasting kurallarına uyan tensörlerde kullanılabilir.

Matematiksel olarak geçerli olan matris çarpımı, mul ile yapılamaz.

Bu tür durumlarda **@ **operatörü (**torch.matmul**) kullanılmalıdır.

**Matris Çarpımı Kuralı**

A : (m × n)

B : (n × k)

A @ B : (m × k)


```python
A = torch.rand(5, 3)
B = torch.rand(3, 4)

C = torch.matmul(A, B)

print(C)
#
```

```python
# Bir tensore bir sayı ekleme yönetemi

A = torch.tensor([[1,2,3,0],
                  [5,6,7,0]]);

A.add_(5);

print(A)
# Matrisdeki her elemana 5 ekledi.
```

```python
# Bir tensoru NumPy dizisine çevirme

A = torch.ones(5);
print(A);

B = A.numpy();

print(B);
```

```python
#Bir Numpy dizisini tensor e çevirme

A = np.ones(5);
print(A);

B = torch.from_numpy(A);

print(B);
```
